# EDA Analysis using Python

```python
# importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dt=pd.read_csv("titanic_dataset.csv")
```

```python
dt
```

```python
dt.info()
```

```python
#DISPLAY NO OF ROWS AND COLUMNS
print("Rows:", dt.shape[0])
print("Columns:", dt.shape[1])
```

```python
#SET PASSENGER ID AS INDEX COLUMN
dt.set_index("PassengerId", inplace=True)
```

```python
dt.describe()
```

**CATEGORICAL DATA ANALYSIS**

```python
# USE VALUE COUNT FUNCTION AND PERFROM CATEGORICAL ANALYSIS
print(dt["Survived"].value_counts())
print(dt["Pclass"].value_counts())
print(dt["Sex"].value_counts())
print(dt["Embarked"].value_counts())    
```

**UNIVARIATE ANALYSIS**

```python
# USE COUNTPLOT AND PERFORM UNIVARIATE ANALYSIS FOR THE "SURVIVED" COLUMN IN TITANIC DATASET
sns.countplot(x="Survived", data=dt)
plt.title("Survival Count")
plt.xlabel("Survived")
plt.ylabel("Count")
plt.show()
```

```python
# IDENTIFY UNIQUE VALUES IN "PASSENGER CLASS" COLUMN
print(dt["Pclass"].unique())
```

```python
# RENAMING COLUMN
dt.rename(columns = {'Sex':'Gender'}, inplace = True)
dt
```

**BIVARIATE ANALYSIS**

```python
# USE CATPLOT METHOD FOR BIVARIATE ANALYSIS
sns.catplot(x="Pclass", hue="Survived", kind="count", data=dt)
plt.title("Survival Count by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(8,5))
graph = sns.countplot(x="Pclass", data=dt)
graph.set_xticklabels(graph.get_xticklabels())
for p in graph.patches:
    height = p.get_height()
    graph.text(p.get_x()+p.get_width()/2, height + 20.8,height ,ha="left")
```

```python
# USE BOXPLOT METHOD TO ANALYZE AGE AND SURVIVED COLUMN
sns.boxplot(x="Survived", y="Age", data=dt)
plt.title("Boxplot of Age by Survival Status")
plt.xlabel("Survived")
plt.ylabel("Age")
plt.show()
```

**MULTIVARIATE ANALYSIS**

```python
# USE BOXPLOT METHOD AND ANALYZE THREE COLUMNS(PCLASS,AGE,GENDER)
sns.boxplot(x="Pclass", y="Age", hue="Gender", data=dt)
plt.title("Boxplot of Age by Passenger Class and Gender")
plt.xlabel("Passenger Class")
plt.ylabel("Age")
plt.legend(title="Gender    ")
plt.show()
```

```python
# USE CATPLOT METHOD AND ANALYZE THREE COLUMNS(PCLASS,SURVIVED,GENDER)
sns.catplot(x="Pclass", hue="Survived", col="Gender", kind="count", data=dt)
plt.subplots_adjust(top=0.9)
plt.suptitle("Survival Count by Passenger Class and Gender")
plt.show()
```

```python
# IMPLEMENT HEATMAP AND PAIRPLOT FOR THE DATASET
sns.heatmap(dt.select_dtypes(include=[np.number]).corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
